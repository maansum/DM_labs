```python
import csv

data = """Bachelors Mobile Dev Objective-C TRUE Yes
Masters Web Dev Java FALSE Yes
Masters Mobile Dev Java TRUE Yes
PhD Mobile Dev Objective-C TRUE Yes
PhD Web Dev Objective-C TRUE No
Bachelors UX Design Objective-C TRUE No
Bachelors Mobile Dev Java FALSE Yes
PhD Web Dev Objective-C FALSE No
Bachelors UX Design Java FALSE Yes
Masters UX Design Objective-C TRUE No
Masters UX Design Java FALSE Yes
PhD Mobile Dev Java FALSE No
Masters Mobile Dev Java TRUE Yes
Bachelors Web Dev Objective-C FALSE No"""

lines = data.split('\n')

rows = []
for line in lines:
    parts = line.split()
    qualification = parts[0]
    course = parts[1]
    language = parts[3]
    is_true = parts[4]
    answer = parts[5]

    rows.append([qualification, course, language, is_true, answer])

# Write data to a single CSV file
with open('combined_data.csv', 'w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Qualification', 'Course', 'Language', 'Is True', 'Answer'])
    writer.writerows(rows)

print("CSV file created successfully.")
```

```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('combined_data.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
qualification=df['Qualification'].values.tolist()
course=df['Course'].values.tolist()
language=df['Language'].values.tolist()
isTrue=df['Is True'].values.tolist()
answer=df['Answer'].values.tolist()
```

```python
encoder=LabelEncoder()
```

```python
qualiEncoder=encoder.fit_transform(qualification)
courseEncoder=encoder.fit_transform(course)
langEncoder=encoder.fit_transform(language)
isTrueEncoder=encoder.fit_transform(isTrue)
ansEncoder=encoder.fit_transform(answer)
```

```python
features = []
for i in range(len(qualiEncoder)):
  features.append([qualiEncoder[i], courseEncoder[i],
  langEncoder[i],isTrueEncoder[i]])
```

```python
classifier=DecisionTreeClassifier(criterion='entropy')
```

```python
classifier=classifier.fit(features,ansEncoder)
```
