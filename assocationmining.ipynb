```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, fpmax, fpgrowth
```

```python
# dataset=[['Milk', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
#            ['Dill', 'Onion', 'Nutmeg', 'Kidney Beans', 'Eggs', 'Yogurt'],
#            ['Milk', 'Apple', 'Kidney Beans', 'Eggs'],
#            ['Milk', 'Unicorn', 'Corn', 'Kidney Beans', 'Yogurt'],
#            ['Corn', 'Onion', 'Onion', 'Kidney Beans', 'Ice cream', 'Eggs']]
data= pd.read_csv('GroceryStoreDataSet.csv').values.tolist()
```

```python
data
```

```python
trans=TransactionEncoder()
transform_arr=trans.fit(data).transform(data)
df=pd.DataFrame(transform_arr, columns=trans.columns_)
```

```python
df
```

```python
frequent_itemssets= fpgrowth(df, min_support=0.1,use_colnames=True)
### alternatively:
#frequent_itemssets = apriori(df, min_support=0.1, use_colnames=True)
#frequent_itemsets = fpmax(df, min_support=0.6, use_colnames=True)
```

```python
frequent_itemssets
```

```python
from mlxtend.frequent_patterns import association_rules
```

```python
rules= association_rules(frequent_itemssets, metric='confidence', min_threshold=0.7)
```

```python
rules
```

```python
support_confidence=rules[['antecedents','consequents', 'support', 'confidence'
                          ]]
support_confidence
```
